```python
X = X.drop(["Dew point temperature(C)"], axis=1)


x_train, x_test, y_train, y_test = tts(X,Y,random_state=1, test_size=Q, shuffle=False)

x_train=sm.add_constant(x_train)
x_test=sm.add_constant(x_test)



DemandaBicicleta=sm.OLS(y_train,x_train).fit()
print(DemandaBicicleta.summary())

```

Calculando metricas

```python
y_pred = DemandaBicicleta.predict(x_test)

y_pred

MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)

print(f"MAE {MAE}")
print(f"MSE {MSE}")
print(f"RMSE {RMSE}")
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.1)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Valores Reales vs. Predicciones')

min_val = 0
max_val = max(max(y_test), max(y_pred))
plt.plot([0, 2500], [0 , 2500], color='red', linestyle='--')

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```

Existe multicolinealidad?

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Añadir constante a X para calcular VIF
X_with_const = sm.add_constant(X)

# Calcular VIF para cada variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]

print(vif_data)
```
